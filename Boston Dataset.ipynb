```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
```

```python
scaler = preprocessing.StandardScaler()
```

```python
traning_data = pd.read_csv('Datasets/Train.csv')
X_test = pd.read_csv('Datasets/Test.csv')
```

```python
X_train = traning_data.iloc[:,:-1]
X_test.columns = X_train.columns
columns = X_train.columns
for i in range(len(columns)):
    for j in range(i,len(columns)):
        X_train[ columns[i]+"_"+columns[j] ] = X_train[columns[i]] * X_train[columns[j]]
X_train
```

```python
columns = X_test.columns
for i in range(len(columns)):
    for j in range(i,len(columns)):
        X_test[ columns[i]+"_"+columns[j] ] = X_test[columns[i]] * X_test[columns[j]]
X_test
```

```python
Y_train = traning_data.iloc[:,-1]
```

```python
Y_train
```

```python
scaler.fit(X_train)
```

```python
X_test = scaler.transform(X_test)
X_train = scaler.transform(X_train)
```

```python
alg = LinearRegression()
alg.fit(X_train,Y_train)
```
